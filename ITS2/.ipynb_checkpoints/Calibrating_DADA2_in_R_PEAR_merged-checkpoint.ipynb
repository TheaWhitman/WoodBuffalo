## Running DADA2 outside of QIIME2 for more flexibility

Largely from DADA2 tutorial: https://benjjneb.github.io/dada2/tutorial.html

```r
library(dada2)
```

```r
library(reshape)
library(ggplot2)
```

```r
path = "../../data/Seq_data/ITS2/PEAR_assembled/assembled" # directory containing the fastq files after unzipping.
list.files(path)

# Sort ensures forward/reverse reads are in same order
fns = sort(list.files(path, pattern="assembled.fastq"))

# Extract sample names, assuming filenames have format: SAMPLENAME_XXX.fastq
sample.names = sapply(strsplit(fns, "_"), `[`, 1)

# Specify the full path to the fnFs and fnRs
fns = file.path(path, fns)
```

```r
fns
```

```r
fnsmini=fns[1:6]
sample.namesmini=sample.names[1:6]
filtsmini = file.path(filt_path, paste0(sample.namesmini, "assembled.fastq"))
```

```r
plotQualityProfile(fns[1:6])
# Looks like you'd about expect - some dropoff in sequence quality for the reverse reads, 
# but certainly much better than just the fwd and reve reads alone
```

```r
filt_path = file.path(path, "filtered") # Place filtered files in filtered/ subdirectory
filts = file.path(filt_path, paste0(sample.names, "assembled.fastq"))
```

```r
out = filterAndTrim(fnsmini, filtsmini, trimLeft=c(0,0), truncLen=c(0,0)
                    ,maxN=0, maxEE=2, truncQ=2, rm.phix=FALSE, n=1e5
                    ,verbose=TRUE ,compress=TRUE, multithread=TRUE)
head(out)
```

```r
out = filterAndTrim(fnsmini, filtsmini, trimLeft=c(0,0), truncLen=c(0,0)
                    ,maxN=0, maxEE=1, truncQ=2, rm.phix=FALSE, n=1e5
                    ,verbose=TRUE ,compress=TRUE, multithread=TRUE)
head(out)
# MaxEE=1 retains ~99% of the reads, and MaxEE=2 retains ~99.9% of the reads.
```

```r
err = learnErrors(filtsmini, nreads=3e+06, randomize=TRUE, multithread=TRUE)
```

```r
plotErrors(err, nominalQ=TRUE)
# These PEAR merged sequences match the expected error profile much better than the unmerged.
```

```r
dereps = derepFastq(filtsmini, 3e+06, verbose=TRUE)
# Dereplicating the sequences, keeping q scores
# Increasing the # seqs processed at any given time just to speed it up

# Name the derep-class objects by the sample names
names(dereps) = sample.namesmini

# This also works much better than with the unmerged reads.
```

```r
dadas = dada(dereps, err=err, selfConsist=FALSE, pool=TRUE
     ,OMEGA_A = 1e-2, USE_QUALS=TRUE,BAND_SIZE=32,VERBOSE=FALSE, multithread=TRUE)
# Self-consist is false because it happens in the error estimation step already.
# https://bmcbioinformatics.biomedcentral.com/articles/10.1186/1471-2105-13-283
# They say Omega_A values of as low as 0.01 can be reasonable on diverse, undersampled data
# Pool is probably good for sparse data like this multi-community ITS2 data.
dadas[[1]]
```

```r
dadas = dada(dereps, err=err, selfConsist=FALSE, pool=TRUE
     ,OMEGA_A = 1e-3, USE_QUALS=TRUE,BAND_SIZE=32,VERBOSE=FALSE, multithread=TRUE)
# Self-consist is false because it happens in the error estimation step already.
# https://bmcbioinformatics.biomedcentral.com/articles/10.1186/1471-2105-13-283
# They say Omega_A values of as low as 0.01 can be reasonable on diverse, undersampled data
# Pool is probably good for sparse data like this multi-community ITS2 data.
dadas[[1]]
# Comparing 0.01 to 0.001 to see if it affects estimation of chimeras
```

```r
dadas = dada(dereps, err=err, selfConsist=FALSE, pool=TRUE
     ,OMEGA_A = 1e-4, USE_QUALS=TRUE,BAND_SIZE=32,VERBOSE=FALSE, multithread=TRUE)
# Self-consist is false because it happens in the error estimation step already.
# https://bmcbioinformatics.biomedcentral.com/articles/10.1186/1471-2105-13-283
# They say Omega_A values of as low as 0.01 can be reasonable on diverse, undersampled data
# Pool is probably good for sparse data like this multi-community ITS2 data.
dadas[[1]]
# Comparing 0.01 to 0.001 to see if it affects estimation of chimeras
```

```r
OTUtab = makeSequenceTable(dadas)
# Make OTU table
dim(OTUtab)
# Samples and OTUs
```

```r
OTUtab.nochim = removeBimeraDenovo(OTUtab, method="consensus", multithread=TRUE, verbose=TRUE)
# Removing chimeras - reiterating after first dada step, based on merged seqs.
dim(OTUtab.nochim)
# 355 chimeras / 906 otus for omega_A 0.01 - 551 remaining
# 329 chimeras / 873 otus for omega_A 0.001 -25 fewer chimeras, -33 fewer otus - 544 remaining
# 298 chimeras / 830 otus for omega_A 0.0001 -31 fewer chimeras, -43 fewer otus - 532 remaining
```

```r
percent = round((1-sum(OTUtab.nochim)/sum(OTUtab))*100,1)
paste(percent,"% of sequences were IDd as chimeras")
# % of sequences that were chimeras
# 8.5% with 11:16 under omega_a 0.01
# 9.1% with 1:6 under omega_a 0.01
# 8.9% with 1:6 under omega_a 0.001 - similar fraction of sequences kept...
# 9.2% with 1:6 under omega_a 0.0001 - similar fraction of sequences kept...

# Would be good to compare to USEARCH chimera detection (suggests dada2 high false pos)
# Will save in CHTC script and compare afterward.
```

```r
# 1:6 under omega_a 0.0001
getN = function(x) sum(getUniques(x))
track = cbind(out, sapply(dadas, getN), rowSums(OTUtab), rowSums(OTUtab.nochim))
colnames(track) = c("input", "filtered", "denoised","tabled", "nonchim")
rownames(track) = sample.namesmini
head(track)
# Seems like we actually kept more seqs under a lower omega_A
```

```r
# 1:6 under omega_a 0.001
getN = function(x) sum(getUniques(x))
track = cbind(out, sapply(dadas, getN), rowSums(OTUtab), rowSums(OTUtab.nochim))
colnames(track) = c("input", "filtered", "denoised","tabled", "nonchim")
rownames(track) = sample.namesmini
head(track)
# Seems like we actually kept more seqs under a lower omega_A after chimera removal
# Perhaps sequences are getting thrown out as chimeras, but if we had... fewer otus
# to begin with, fewer otus get IDd as chimeras? Perhaps that is it.
    
# I think I will just try running CHTC with dada2 omega_a 1e-20 and v liberal (0.01)
```

```r
# 9.1% with 1:6 under omega_a 0.01
getN = function(x) sum(getUniques(x))
track = cbind(out, sapply(dadas, getN), rowSums(OTUtab), rowSums(OTUtab.nochim))
colnames(track) = c("input", "filtered", "denoised","tabled", "nonchim")
rownames(track) = sample.namesmini
head(track)
```

```r
#Previous results with 11:16 under omega_a 0.01
getN = function(x) sum(getUniques(x))
track = cbind(out, sapply(dadas, getN), rowSums(OTUtab), rowSums(OTUtab.nochim))
colnames(track) = c("input", "filtered", "denoised","tabled", "nonchim")
rownames(track) = sample.namesmini
head(track)
```

```r
options(repr.plot.width=6, repr.plot.height=3)
df = data.frame(melt(track))
p = ggplot(df,aes(y=value,x=X2, color=X1))
p = p + geom_point()
p = p + ylim(0,24000)
p
# We can see a step-change at the merging step.
# Also note these popped out in the "wrong" order. (input >> nochim)
```
