## Making plots to compare to the Holden et al. burn severity paper
"We calculated the ‘fire severity tolerance’ for each taxon as the average dNBR of the sites in which it was observed. We computed fire severity tolerance based on (1) presence/absence data and (2) weighted by the relative abundance of the taxon at the sites in which it was observed. Although these values were similar for most fungal taxa, to be conservative we present the fire severity tolerance based on the presence/absence data."

```r
# Loading required packages
library(reshape)
library(ggplot2)
library(phyloseq)
library(dplyr)
library(wesanderson)
```

```r
ps = readRDS(file="CHTC/Dada2_Results_OA20_PEAR_Full/ps.fungi")
ps.norm = transform_sample_counts(ps, function(x) (x / sum(x)))
d = psmelt(ps.norm)
```

```r
HoldenData = data.frame(Phylum=c("Ascomycota","Basidiomycota","Chytridiomycota","Glomeromycota","Mortierellomycota","Mucoromycota","Rozellomycota"), MeanRBR = c(580,410,550,440,585,750,NA))
HoldenData
```

```r
# This approach groups by each OTU
df = d %>%
    filter(!(Phylum %in% c("","unidentified")))%>%
#    filter(Severity_Class!="Unburned")%>%
    filter(Land_Class != "Wetland")%>%
#    group_by(Land_Class,Org_or_Min,Phylum,Class,Order,Family,Genus,Species,OTU,CBI,RBR,Burn_Severity_Index)%>%
    group_by(Site_ID,Veg_Comm,Phylum,Class,Order,Family,Genus,Species,OTU,CBI,RBR,Burn_Severity_Index)%>%
    summarize(TaxPresent = ifelse(sum(Abundance)>0,1,0))%>%
    filter(TaxPresent>0)%>%
#    group_by(Veg_Comm,Phylum,Class,Order,Family,Genus,Species,OTU)%>%
    group_by(Phylum,Class,Order,Family,Genus,Species,OTU)%>%
    summarize(MeanCBI=mean(CBI),MeanRBR=mean(RBR),MeanBSI=mean(Burn_Severity_Index))
# This should have taken all samples that are not wetlands or unburned
# And then, it filters out all taxa instances where it is not present (Abundance=0)
# Then, it takes the average burn severity (using different metrics) for each OTU.
# It hangs on to the other information, so we can plot it by phylum, etc.
# It looks for presence/absence at each site, so we don't treat org or min differently.
```

```r
options(repr.plot.width=3, repr.plot.height=3)

p = ggplot(df)

p = ggplot(df, aes(x=Phylum, y=MeanRBR, colour=Phylum))
p = p + geom_boxplot()
#p = p + geom_jitter(alpha=0.05)
#p = p + geom_point(data = HoldenData, mapping = aes(x = Phylum, y = MeanRBR), colour="black", size=3, shape=4)
#p = p + facet_wrap(~Veg_Comm, scales="free")
p = p + theme_bw()
p = p + theme(axis.text.x = element_text(angle=45,hjust=1, face="italic"))
p = p + guides(colour=FALSE) + expand_limits(y=0)
#p = p + geom_text(aes(label=paste("",nTax, sep=""), y = 3.1, angle=90), colour="black")
p
```

```r
a = aov(MeanRBR~Phylum, data=df)
Tukey = TukeyHSD(a)
summary(a)
Tukey
```

## Considering veg comms separately

```r
# This approach groups by each OTU
df = d %>%
    filter(!(Phylum %in% c("","unidentified")))%>%
    filter(Severity_Class!="Unburned")%>%
    filter(Land_Class != "Wetland")%>%
    group_by(Site_ID,Veg_Comm,Phylum,Class,Order,Family,Genus,Species,OTU,CBI,RBR,Burn_Severity_Index)%>%
    summarize(TaxPresent = ifelse(sum(Abundance)>0,1,0))%>%
    filter(TaxPresent>0)%>%
    group_by(Veg_Comm,Phylum,Class,Order,Family,Genus,Species,OTU)%>%
    summarize(MeanCBI=mean(CBI),MeanRBR=mean(RBR),MeanBSI=mean(Burn_Severity_Index))
# This should have taken all samples that are not wetlands or unburned
# And then, it filters out all taxa instances where it is not present (Abundance=0)
# Then, it takes the average burn severity (using different metrics) for each OTU.
# It hangs on to the other information, so we can plot it by phylum, etc.
# It looks for presence/absence at each site, so we don't treat org or min differently.
```

```r
options(repr.plot.width=10, repr.plot.height=3)

p = ggplot(df)

p = ggplot(df, aes(x=Phylum, y=MeanRBR, colour=Phylum))
p = p + geom_boxplot()
p = p + geom_point(data = HoldenData, mapping = aes(x = Phylum, y = MeanRBR), colour="black", size=3, shape=4)
p = p + facet_wrap(~Veg_Comm, scales="free")
p = p + theme_bw()
p = p + theme(axis.text.x = element_text(angle=45,hjust=1, face="italic"))
p = p + guides(colour=FALSE) + expand_limits(y=0)
#p = p + geom_text(aes(label=paste("",nTax, sep=""), y = 3.1, angle=90), colour="black")
p
```

## Considering weighted meanRBRs

```r
# This approach groups by each OTU
df = d %>%
    filter(!(Phylum %in% c("","unidentified")))%>%
    filter(Severity_Class!="Unburned")%>%
    filter(Land_Class != "Wetland")%>%
    group_by(Org_or_Min,Phylum,Class,Order,Family,Genus,Species,OTU,CBI,RBR,Burn_Severity_Index,Abundance)%>%
    summarize(MeanCBI=weighted.mean(CBI,Abundance),MeanRBR=weighted.mean(RBR,Abundance),MeanBSI=weighted.mean(Burn_Severity_Index,Abundance))
# This should have taken all samples that are not wetlands or unburned
# Then, it takes the weighted average of burn severity (using different metrics) for each OTU.
# It does this separately for organic vs. mineral horizons.
# It hangs on to the other information, so we can plot it by phylum, etc.
head(df)
```

```r
options(repr.plot.width=3, repr.plot.height=3)

p = ggplot(df)

p = ggplot(df, aes(x=Phylum, y=MeanRBR, colour=Phylum))
p = p + geom_boxplot()
p = p + geom_point(data = HoldenData, mapping = aes(x = Phylum, y = MeanRBR), colour="black", size=3, shape=4)
#p = p + facet_wrap(~Veg_Comm, scales="free")
p = p + theme_bw()
p = p + theme(axis.text.x = element_text(angle=45,hjust=1, face="italic"))
p = p + guides(colour=FALSE) + expand_limits(y=0)
#p = p + geom_text(aes(label=paste("",nTax, sep=""), y = 3.1, angle=90), colour="black")
p
```
