## Running DADA2 outside of QIIME2 for more flexibility

Largely from DADA2 tutorial: https://benjjneb.github.io/dada2/tutorial.html

```r
library(dada2)
```

```r
path = "../data/Seq_data/ITS2"
list.files(path)

# Sort ensures forward/reverse reads are in same order
fnFs = sort(list.files(path, pattern="_R1_001.fastq.gz"))
fnRs = sort(list.files(path, pattern="_R2_001.fastq.gz"))
# Extract sample names, assuming filenames have format: SAMPLENAME_XXX.fastq
sample.names = sapply(strsplit(fnFs, "_"), `[`, 1)
# Specify the full path to the fnFs and fnRs
fnFs = file.path(path, fnFs)
fnRs = file.path(path, fnRs)
```

```r
plotQualityProfile(fnFs[1:6])
```

```r
filt_path = file.path(path, "filtered") # Place filtered files in filtered/ subdirectory
filtFs = file.path(filt_path, paste0(sample.names, "_F_filt.fastq.gz"))
filtRs = file.path(filt_path, paste0(sample.names, "_R_filt.fastq.gz"))
```

```r
out = filterAndTrim(fnFs, filtFs, fnRs, filtRs, trimLeft=c(0,0), truncLen=c(0,0)
                    ,maxN=0, maxEE=c(5,5), truncQ=2, rm.phix=TRUE, n=1000000
                    ,verbose=TRUE ,compress=TRUE, multithread=TRUE)
head(out)
```

```r
errF = learnErrors(filtFs, nreads=1e+06, randomize=TRUE, multithread=TRUE)
errR = learnErrors(filtRs, nreads=1e+06, randomize=TRUE, multithread=TRUE)
```

```r
plotErrors(errR, nominalQ=TRUE)
# In the .mini dataset, the error estimates weren't that good - would have needed more seqs.
```

```r
plotErrors(errF, nominalQ=TRUE)
# In the .mini dataset, the error estimates weren't that good - would have needed more seqs.
```

```r
derepFs = derepFastq(filtFs, 3e+06)
derepRs = derepFastq(filtRs, 3e+06)
# Dereplicating the sequences, keeping q scores
# Increasing the # seqs processed at any given time just to speed it up

# Name the derep-class objects by the sample names
names(derepFs) = sample.names
names(derepRs) = sample.names
```

```r
# Need to test OMEGA_A, and pooling.
```

```r
#dada(derep1, err=inflateErr(tperr1,3), BAND_SIZE=32, OMEGA_A=1e-20)
dadaFs = dada(derepFs, err=errF, selfConsist=FALSE, pool=TRUE
     ,OMEGA_A = 1e-2, USE_QUALS=TRUE,BAND_SIZE=32,VERBOSE=FALSE, multithread=TRUE)
# Self-consist is false because it happens in the error estimation step already.
# https://bmcbioinformatics.biomedcentral.com/articles/10.1186/1471-2105-13-283
# They say Omega_A values of as low as 0.01 can be reasonable on diverse, undersampled data, depending on tests
# Will need to test further, but for now this seems like an ok approach for at least the first pass.
dadaFs[[1]]
```

```r
mergers = mergePairs(dadaFs, derepFs, dadaRs, derepRs, minOverlap=20, maxMismatch=2, verbose=TRUE)
OTUtab = makeSequenceTable(mergers)
# Make OTU table
dim(OTUtab)
# Samples and OTUs
```

```r
OTUtab.nochim = removeBimeraDenovo(OTUtab, method="consensus", multithread=TRUE, verbose=TRUE)
# Removing chimeras - reiterating after first dada step, based on merged seqs.
dim(OTUtab.nochim)
sum(OTUtab.nochim)/sum(OTUtab)
```

```r
getN = function(x) sum(getUniques(x))
track = cbind(out, sapply(dadaFs, getN), sapply(mergers, getN), rowSums(OTUtab), rowSums(OTUtab.nochim))
colnames(track) = c("input", "filtered", "denoised", "merged", "tabled", "nonchim")
rownames(track) = sample.names
head(track)
```

```r
options(repr.plot.width=6, repr.plot.height=3)
df = data.frame(melt(track))
p = ggplot(df,aes(y=value,x=X2, color=X1))
p = p + geom_point()
p = p + ylim(0,24000)
p
# We can see a step-change at the merging step.
```
