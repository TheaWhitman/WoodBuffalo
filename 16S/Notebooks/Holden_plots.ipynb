```r
# Loading R packages
library(reshape)
library(ggplot2)
library(phyloseq)
library(plyr)
library(dplyr)
library(plotly)
library(wesanderson)
library(agricolae)
library(vegan)
```

```r
# Importing the dataset
ps = readRDS("../../../data/Seq_data/16S/CHTC/Dada2_Results/ps.merged")
ps.norm = transform_sample_counts(ps, function(x) (x / sum(x)))
```

```r
mdf = psmelt(ps.norm)
```

```r
df = mdf %>%
    group_by(Veg_Comm,Org_or_Min,Phylum,Sample,OTU,CBI,RBR,Burn_Severity_Index)%>%
    summarize(TaxPresent = ifelse(sum(Abundance)>0,1,0))%>%
    filter(TaxPresent>0)%>%
    filter(Phylum != "unidentified")%>%
    filter(Phylum != "")%>%
    group_by(Veg_Comm,Org_or_Min,Phylum,OTU)%>%
    summarize(MeanCBI=mean(CBI),MeanRBR=mean(RBR),MeanBSI=mean(Burn_Severity_Index),nTax=n())%>%
    group_by(Veg_Comm,Org_or_Min,Phylum)%>%
    mutate(sdCBI=sd(MeanCBI,na.rm=TRUE),sdRBR=sd(MeanRBR,na.rm=TRUE),sdBSI=sd(MeanBSI,na.rm=TRUE))%>%
    summarize(MeanCBI=mean(MeanCBI),MeanRBR=mean(MeanRBR),MeanBSI=mean(MeanBSI),sdCBI=mean(sdCBI),sdRBR=mean(sdRBR),sdBSI=mean(sdBSI),nTax=sum(nTax))%>%
    mutate(seCBI=sdCBI/(nTax^0.5),seRBR=sdRBR/(nTax^0.5),seBSI=sdRBR/(nTax^0.5))

Order = df %>%
    group_by(Phylum)%>%
    summarize(Order=mean(MeanRBR))%>%
    arrange(-Order)
```

```r
options(repr.plot.width=8, repr.plot.height=6)

df$Phylum = ordered(df$Phylum, levels=paste(Order$Phylum))

df = df%>%
    filter(Phylum != "unidentified")%>%
    filter(Phylum != "")

p = ggplot(df, aes(x=Phylum, y=MeanRBR, fill=Phylum))
p = p + geom_col()
p = p + geom_errorbar(data=df, aes(ymin=MeanRBR-1.96*seRBR, ymax=MeanRBR+1.96*seRBR), color="black")
p = p + facet_grid(~Veg_Comm~Org_or_Min, scales="free")
p = p + theme(axis.text.x = element_text(angle=45,hjust=1))
p = p + guides(fill=FALSE)
p = p + ylim(0,750)
p = p + geom_text(aes(label=paste("",nTax, sep=""), y = 700, angle=90), colour="black", size=2.5)
p = p + theme(legend.position="bottom")
p
# Need to think about exactly what we want to calculate here...
```

```r
df = mdf %>%
    group_by(Veg_Comm,Org_or_Min,Class,OTU,CBI,RBR,Burn_Severity_Index)%>%
    summarize(TaxPresent = ifelse(sum(Abundance)>0,1,0))%>%
    filter(TaxPresent>0)%>%
    filter(Class != "unidentified")%>%
    filter(Class != "")%>%
    group_by(Veg_Comm,Org_or_Min,Class,OTU)%>%
    summarize(MeanCBI=mean(CBI),MeanRBR=mean(RBR),MeanBSI=mean(Burn_Severity_Index),nTax=n())%>%
    group_by(Veg_Comm,Org_or_Min,Class)%>%
    mutate(sdCBI=sd(MeanCBI,na.rm=TRUE),sdRBR=sd(MeanRBR,na.rm=TRUE),sdBSI=sd(MeanBSI,na.rm=TRUE))%>%
    summarize(MeanCBI=mean(MeanCBI),MeanRBR=mean(MeanRBR),MeanBSI=mean(MeanBSI),sdCBI=mean(sdCBI),sdRBR=mean(sdRBR),sdBSI=mean(sdBSI),nTax=sum(nTax))%>%
    mutate(seCBI=sdCBI/(nTax^0.5),seRBR=sdRBR/(nTax^0.5),seBSI=sdRBR/(nTax^0.5))

Order = df %>%
    group_by(Class)%>%
    summarize(Order=mean(MeanRBR))%>%
    arrange(-Order)
```

```r
model = lm(data=df, MeanCBI~Phylum)
anova = aov(model)
summary(anova)
TukeyHSD = TukeyHSD(x=anova,"Phylum")
TukeyHSD
```

```r
test = HSD.test(anova, "Phylum", group=TRUE)
```

```r
modelfunction = function(df,Land_Class, Org_or_Min){
    d = df[df$Land_Class==Land_Class,]
    d = d[d$Org_or_Min==Org_or_Min,]
    #print(length(levels(droplevels(d$Phylum))))
    model = lm(data=d, MeanRBR~Phylum)
    anova = aov(model)
    test = TukeyHSD(anova, "Phylum")
    test
}
```

```r
results = modelfunction(df,c("Upland","Wetland"),c("M","O"))
```
